```python
from bs4 import BeautifulSoup
import requests
url = "https://www.immoweb.be/en/search/house/for-rent?countries=BE&page=1&orderBy=relevance/"
r = requests.get(url)

#print(url, r.status_code)
soup = BeautifulSoup(r.content, "lxml")
soup
for elem in soup.find_all( attrs={"class": "main"}):
    print(elem.get("title"))
```

```python
import pandas as pd
from pandas import DataFrame
links = []
for elem in soup.find_all("a", attrs={"class": "meta"}):
    links.append(elem.get("href"))
links


df1 = pd.DataFrame(data= links,columns ={"linkss"})
#df["Synopsis"] = synopsis
df1
#df.to_csv("./assets/allo_cine.csv", index=False)
```
